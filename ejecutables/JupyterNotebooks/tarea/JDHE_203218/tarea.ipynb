Paso 4. Crea dos nuevas columnas en el gdf original que sean para el area y el perimetro, en ellas haz tu calculo de cada una.

```python
# Cambiar la proyecciÃ³n del GeoDataFrame a EPSG:32615
df = df.to_crs('EPSG:32615')
#Calcular el area y perimetro
df['area'] = df.geometry.area
df['perimetro'] = df.geometry.length
#Imprimo el gdf
df
```

Paso 5. Crea una nueva columna en el gdf del paso 1 que se llame distancia; despuÃ©s escoge una escuela del data frame y que la nueva columna tenga la distancia de todas las escuelas a la que escogiste 

```python
#Seleccionar la coordenada de la escuela
punto = df[df['nombre']=='ARMONIA INFANTIL'].iloc[0]['geometry']

#Calculo la distancia con respecto al punto e imprimo el gdf
df['distancia'] = df['geometry'].distance(punto)
df
```

Paso 6. Realiza una consulta para saber cuales son las 3 escuelas mas lejos del punto que elegiste, debe imprimir solamente el nombre de las escuelas y su distancia al punto

```python
#Ordeno las escuelas de la mas lejana a la mas cercana 
df_ordenado = df.sort_values(by='distancia',ascending=False)
#Imprimo el nombre y la distancia de las 3 mas lejanas
df_ordenado.head(3)[['nombre','distancia']]
```

Paso 7. Agrupa las escuelas por su alcaldia. Una vez hecho eso crea una nueva geometria que "envuelva" a las escuelas de cada alcaldia y graficalos

```python
# Manejar valores nulos en la columna 'alcaldi'
df['alcaldi'].fillna('Desconocida', inplace=True)

# Agrupar las escuelas por alcaldÃ­a y crear una nueva geometrÃ­a que envuelva las escuelas de cada alcaldÃ­a
df_alcaldias = df.groupby('alcaldi')['geometry'].apply(lambda x: gpd.GeoSeries(x).unary_union)

# Crear un nuevo GeoDataFrame con las geometrÃ­as resultantes
gdf_alcaldias = gpd.GeoDataFrame(geometry=df_alcaldias)

# Crear un mapeo de colores para cada alcaldÃ­a
colormap = plt.cm.get_cmap('viridis', len(df_alcaldias))
colors = range(len(df_alcaldias))

# Graficar el resultado asignando colores diferentes a cada alcaldÃ­a
ax1 = df.plot(color='gray', label='Main data')
gdf_alcaldias.plot(ax=ax1, color=colormap(colors), alpha=0.5, legend=True)

# Configurar el aspecto del grÃ¡fico
ax1.set_title('Envoltura de Escuelas por AlcaldÃ­a')
ax1.set_xlabel('Longitud')
ax1.set_ylabel('Latitud')

# Mostrar el grÃ¡fico
plt.show()
```

El paso anterior guardalo como una capa y graficalo en conjunto a la grafica del paso 1

```python
#Guardo el gdf en un .geojson
gdf_alcaldias.to_file('alcaldias_envelope.geojson', driver='GeoJSON')
#Grafico
ax2 = df.plot(color='gray', label='Main data')
gdf_alcaldias.plot(ax=ax2, color=colormap(colors), alpha=0.5, legend=True)
```
