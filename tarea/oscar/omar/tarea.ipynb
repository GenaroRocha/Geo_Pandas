Paso 5. Crea una nueva columna en el gdf del paso 1 que se llame distancia; despuÃ©s escoge una escuela del data frame y que la nueva columna tenga la distancia de todas las escuelas a la que escogiste 

```python

# Seleccionar una escuela como punto de referencia
# En este caso, he elegido la primera escuela en el DataFrame
escuela_referencia = df_escuelas.geometry.iloc[0]

# Calcular la distancia de cada escuela a la escuela de referencia
df_escuelas['distancia'] = df_escuelas.geometry.distance(escuela_referencia)

# Imprimir el DataFrame para verificar la nueva columna
df_escuelas
```

Paso 7. Agrupa las escuelas por su alcaldia. Una vez hecho eso crea una nueva geometria que "envuelva" a las escuelas de cada alcaldia y graficalos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que df_escuelas es tu GeoDataFrame y ya estÃ¡ cargado con los datos

# Agrupar las escuelas por alcaldÃ­a
grupos = df_escuelas.groupby('alcaldi')

# Crear un nuevo GeoDataFrame para los convex hulls
hulls = gpd.GeoDataFrame(columns=['alcaldi', 'geometry'])

# Calcular el convex hull de cada grupo y agregarlo al GeoDataFrame 'hulls'
for name, group in grupos:
    hull = group.geometry.unary_union.convex_hull
    hulls = pd.concat([hulls, gpd.GeoDataFrame([[name, hull]], columns=['alcaldi', 'geometry'])], ignore_index=True)

# Configurar el CRS (Sistema de Coordenadas de Referencia) para el nuevo GeoDataFrame
# AsegÃºrate de utilizar el CRS adecuado para tus datos
hulls.set_crs(df_escuelas.crs, inplace=True)

# Graficar
hulls.plot()
plt.show()



```
